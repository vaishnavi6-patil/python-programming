```python
#Integer
a = 10
print(a)
print(type(a))

# Float
mark = 84.50
print(mark)
print( type(mark))

# String
name = "vaishnavi"
print( name)
print( type(name))

# Boolean
b = True
print( b)
print(type(b))

# List
x = [1,2,3,4,5]
print(x)
print( type(x))

# Tuple
numbers = (1, 2, 3, 4)
print( numbers)
print( type(numbers))

# Dictionary
student = {
    "Name": "vaishnavi",
    "Age": 20,
    "Course": "Engineering"
}
print(student)
print( type(student))

# Set
colors = {"Red", "Green", "Blue"}
print( colors)
print(type(colors))
```
