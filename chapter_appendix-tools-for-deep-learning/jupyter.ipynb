```python

# 1. Find greatest of 3 numbers
a = 26
b = 39
c = 31

if a > b and a > c:
    print(a)
elif b > a and b > c:
    print(b)
else:
    print(c)


# 2. Odd or Even Number
x = 33

if x % 2 == 0:
    print("even number")
else:
    print("odd number")


# 3. Uppercase or Lowercase
ch = 'V'

if ch.isupper():
    print("uppercase letter")
else:
    print("lowercase letter")


# 4. Number or Character
p = '2'

if p.isdigit():
    print("It is a number")
else:
    print("It is a character")
```
